```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
car_df = pd.read_csv('cars.csv')
```

```python
brand_df = pd.read_csv('brand.csv')
```

```python
car_df.head()
```

```python
brand_df.head()
```

```python
1. cars.csv와 brand.csv의 데이터 합치기
```

```python
# title에서 브랜드 이름 추출하기

car_df['brand'] = car_df['title'].str.split(' ').str[0].str.lower()
```

```python
# 브랜드 이름을 기준으로 두 데이터프레임 병합하기

merged_df = pd.merge(car_df, brand_df, how='left', left_on='brand', right_on='title')
```

```python
merged_df.head()
```

```python
# 불필요한 'title' 열 제거하기

merged_df = merged_df.drop(columns=['title_y'])
```

```python
# 열 이름 변경하기

merged_df = merged_df.rename(columns={'title_x': 'Car Title', 'brand': 'Brand', 'country': 'Country of Origin'})
```

```python
merged_df.head()
```

```python
2. 카테고리형 변수를 숫자 형태로 변환하기
```

```python
# Engine 변수 변환하기

engine_dummies = pd.get_dummies(merged_df['Engine'], prefix='Engine')
```

```python
# Emission Class 변수 변환하기

emission_dummies = pd.get_dummies(merged_df['Emission Class'], prefix='Emission Class')
```

```python
# 원-핫 인코딩된 변수를 원래의 데이터프레임에 추가하기

merged_df = pd.concat([merged_df, engine_dummies, emission_dummies], axis=1)
```

```python
# 원래의 Engine, Emission Class 열 제거하기

merged_df = merged_df.drop(columns=['Engine', 'Emission Class'])
```

```python
merged_df.head()
```

```python
3. 결측치 처리하기
```

```python
# Service history 변수의 결측치를 'Unknown'으로 채우기

merged_df['Service history'].fillna('Unknown', inplace=True)
```

```python
print(merged_df['Service history'])
```

```python
# 결측치가 일정 개수 이상 포함된 데이터 제거하기 (결측치가 2개 이상인 행을 제거)

merged_df.dropna(thresh=merged_df.shape[1]-1, inplace=True)
```

```python
merged_df.head()
```

```python
4. Scaler 활용해보기 
```

```python
# StandardScaler 객체 생성하기

scaler = StandardScaler()
```

```python
5. 기존 데이터의 정보 70% 이상을 가지는 수준에서 최소한의 주성분 추출하기 (PCA)
```

```python
from sklearn.decomposition import PCA
```

```python
# 문자열을 포함하는 열 제거하기

numeric_df = merged_df.select_dtypes(include=[np.number])
```

```python
# PCA 객체 생성하기

pca = PCA(n_components=0.7)
```

```python
# PCA 적용하기 (오류발생)

pca_result = pca.fit_transform(numeric_df)
```

```python
# 오류 해결을 위한 노력하기
```

```python
from sklearn.impute import SimpleImputer
from numpy import inf
```

```python
# 결측치를 평균 값으로 대체하기

imputer = SimpleImputer(strategy='mean')
numeric_df_imputed = imputer.fit_transform(numeric_df)
```

```python
# 무한대 값을 최대 값으로 대체하기

numeric_df_imputed[numeric_df_imputed == inf] = numeric_df_imputed[numeric_df_imputed != inf].max()
```

```python
# PCA 적용하기(**해결됨**)

pca_result = pca.fit_transform(numeric_df_imputed)
```

```python
print("원래 데이터의 차원: ", numeric_df_imputed.shape[1])
print("축소된 데이터의 차원: ", pca_result.shape[1])
```

```python
import matplotlib.pyplot as plt
```

```python
# PCA 결과를 DataFrame으로 변환하기

pca_df = pd.DataFrame(data = pca_result, columns = ['principal component ' + str(i+1) for i in range(pca_result.shape[1])])
```

```python
pca_df.head()
```

```python
# 주성분 값 구하기

pc1 = [-3.766619, 4.608937, 2.880534, 5.525144, 4.716095]
```

```python
# 산점도 그리기

plt.figure(figsize=(8,6))
plt.scatter(range(len(pc1)), pc1)
plt.xlabel('Index')
plt.ylabel('Principal Component 1')
plt.title('Scatter plot of Principal Component 1')
plt.show()
```

```python
7. 상관관계가 높은 변수들이 있는지 확인하기
```

```python
import pandas as pd
```

```python
# 데이터 정리하기

data = {
    'title_x': ['SKODA FABIA', 'VAUXHALL CORSA', 'HYUNDAI I30', 'MINI HATCH', 'VAUXHALL CORSA'],
    'Price': [6900, 1495, 949, 2395, 1000],
    'Mileage(miles)': [70189, 88585, 137000, 96731, 85000],
    'Registration_Year': [2016, 2008, 2011, 2010, 2013],
    'Previous Owners': [3.0, 4.0, None, 5.0, None],
    'Fuel type': ['Diesel', 'Petrol', 'Petrol', 'Petrol', 'Diesel'],
    'Body type': ['Hatchback', 'Hatchback', 'Hatchback', 'Hatchback', 'Hatchback'],
    'Engine': ['1.4L', '1.2L', '1.4L', '1.4L', '1.3L'],
    'Gearbox': ['Manual', 'Manual', 'Manual', 'Manual', 'Manual'],
    'Doors': [5.0, 3.0, 5.0, 3.0, 5.0],
    'Seats': [5.0, 5.0, 5.0, 4.0, 5.0],
    'Emission Class': ['Euro 6', 'Euro 4', 'Euro 5', 'Euro 4', 'Euro 5'],
    'Service history': [None, 'Full', None, 'Full', None],
    'brand': ['skoda', 'vauxhall', 'hyundai', 'mini', 'vauxhall'],
    'title_y': ['skoda', 'vauxhall', 'hyundai', 'mini', 'vauxhall'],
    'country': ['Czech Republic', 'United Kingdom', 'South Korea', 'United Kingdom', 'United Kingdom']
}
```

```python
# 데이터프레임 생성하기

car_df = pd.DataFrame(data)
```

```python
# 모든 변수 간의 상관계수 계산하기

corr_matrix = car_df.corr()
```

```python
# 상관계수가 0.8 이상이고 1 미만인 변수 쌍 찾기

high_corr_var = [(column1, column2) for column1 in corr_matrix.columns for column2 in corr_matrix.columns if (abs(corr_matrix[column1][column2]) > 0.8) & (abs(corr_matrix[column1][column2]) < 1)]
```

```python
for pair in high_corr_var:
    print(f"변수 쌍: {pair}, 상관계수: {corr_matrix[pair[0]][pair[1]]}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 상관계수 행렬을 시각화하기

plt.figure(figsize=(10, 10))
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', cbar=True, square=True)
```
