```python
sigungus_id_dict = {}
current_id = 1

def assign_sigungus_id(sigungus):
    global current_id
    if sigungus not in sigungus_id_dict:
        sigungus_id_dict[sigungus] = current_id
        current_id += 1
    return sigungus_id_dict[sigungus]

df['시군구_ID'] = df['시군구'].apply(assign_sigungus_id)
df
```
